## Team Wise Analysis

```python
# RCB Vs MI
def team1_vs_team2(team1,team2):
    comp=matches[((matches['team1']==team1)|(matches['team2']==team1))&((matches['team1']==team2)|(matches['team2']==team2))]
    sns.countplot(x='season',hue='winner',data=comp)
    plt.xticks(rotation='vertical')
    leg=plt.legend(loc='upper center')
    fig=plt.gcf()
    fig.set_size_inches(10,6)
    plt.grid(axis='y')
    plt.show()
team1_vs_team2('Royal Challengers Bangalore','Mumbai Indians')
```

Total 25 matches played between them, out of that MI won 16 matches and RCB won 9 matches. In 2010 they played qualifier match. so, similar comaprison can be done between any teams.

```python
# MI vs CSK
def team1_vs_team2(team1,team2):
    comp=matches[((matches['team1']==team1)|(matches['team2']==team1)) & ((matches['team1']==team2)|(matches['team2']==team2))]
    sns.countplot(x='season',hue='winner',data=comp)
    plt.xticks(rotation='vertical')
    leg=plt.legend(loc='upper left')
    fig=plt.gcf()
    fig.set_size_inches(10,5)
    plt.grid()
    plt.show()
team1_vs_team2("Mumbai Indians","Chennai Super Kings")
```

In 2016 & 2017 CSK was not in the series, Total 28 matches played between MI & CSK out of MI won 17 matches and CSK won 11 matches.

```python
# let's see winning percentage of teams batting second
num_wins=(matches["win_by_wickets"]>0).sum()
num_loss=(matches["win_by_wickets"]==0).sum()
labels=["Wins","Loss"]
total=float(num_wins+num_loss)
sizes=[(num_wins/total)*100,(num_loss/total)*100]
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90,radius=1.2)
plt.title("Win percentage batting second")
plt.show()
```

Teams batting second has won  53.7%.

```python
#Top players of all season
temp=matches["player_of_match"].value_counts()[:20]
labels=np.array(temp.index)
index=np.arange(len(labels))
width=0.8

fig, ax=plt.subplots(figsize=(15,8))
rects=ax.bar(index,np.array(temp),width=width)
ax.set_xticks(index+((width)/10))
ax.set_xticklabels(labels,rotation='vertical')

for p in ax.patches:
    ax.annotate(format(p.get_height()),(p.get_x()+0.15,p.get_height()+1),fontsize=10)
ax.tick_params(labelsize=10)

plt.title("Top players of the match awardees")


plt.show()
```

CH Gayle is the top player of the match awardee in all IPL season

```python
### Top Umpires
temp=pd.melt(matches, id_vars=['id'],value_vars=['umpire1','umpire2'])
umpire=temp.value.value_counts()[:10]
labels=np.array(umpire.index)
ind=np.arange(len(labels))
width=0.5
fig, ax=plt.subplots(figsize=(15,8))
rects=ax.bar(ind,np.array(umpire),width=width)
ax.set_xticks(ind+((width)/10))
ax.set_xticklabels(labels,rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top 10 Umpires",fontweight='bold',fontfamily='monospace')

for p in ax.patches:
    ax.annotate(format(p.get_height()),(p.get_x()+0.1,p.get_height()+1),fontsize=10)
plt.show()
```

## Batsman Analysis

```python
# Most number of IPL runs 
temp_df=delivery.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
temp_df=temp_df.iloc[:10,:]
labels=np.array(temp_df['batsman'])
ind=np.arange(len(labels))
width=0.5
fig, ax=plt.subplots(figsize=(10,5))
rects=ax.bar(ind,np.array(temp_df['batsman_runs']),width=width,color='orange')
ax.set_xticks((ind)+((width)/10))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("count")
ax.set_title('Top scorer in IPL',fontweight='bold')
ax.set_xlabel('Batsman Name')

for p in ax.patches:
    ax.annotate(format(p.get_height()),(p.get_x()+0.15,p.get_height()+1))

plt.show()
```

Virat Kohli is the leading scorer in IPL. 

```python
# Now let's check most number of 6's
temp_df=delivery.groupby('batsman')['batsman_runs'].agg(lambda x :(x==6).sum()).reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
temp_df=temp_df.iloc[:10,:]
labels=np.array(temp_df['batsman'])
ind=np.arange(len(labels))
width=0.5
fig, ax=plt.subplots(figsize=(10,5))
rects=ax.bar(ind, np.array(temp_df['batsman_runs']),width=width)
ax.set_xticklabels(labels, rotation='vertical')
ax.set_xticks(ind+((width)/10))
ax.set_xlabel('BATSMAN Name')
ax.set_title("Batsman with maximum number of six",fontweight='bold')

for p in ax.patches:
    ax.annotate(format(p.get_height()),(p.get_x()+0.15,p.get_height()+1))
plt.show()
```

CH Gayle is the master in hitting maximum number of sixes.

```python
# Now see the maximum number of 4's
temp_df=delivery.groupby('batsman')['batsman_runs'].agg(lambda x :(x==4).sum()).reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
temp_df=temp_df.iloc[:10,:]
labels=np.array(temp_df['batsman'])
ind=np.arange(len(labels))
width=0.5

fig, ax=plt.subplots(figsize=(10,5))
rects=ax.bar(ind, np.array(temp_df['batsman_runs']),width=width)
ax.set_xticklabels(labels, rotation='vertical')
ax.set_xticks(ind+((width)/10))
ax.set_xlabel('BATSMAN Name')
ax.set_title("Batsman with maximum number of 4's",fontweight='bold')


for p in ax.patches:
    ax.annotate(format(p.get_height()),(p.get_x()+0.15,p.get_height()+1))
plt.show()
```

S.Dhawan is maximum number of 4's hitter.

```python
# Now let us see batman who played most number of dotballs
temp_df=delivery.groupby('batsman')['batsman_runs'].agg(lambda x: (x==0).sum()).reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
temp_df=temp_df.iloc[:10,:]
labels=np.array(temp_df['batsman'])
ind=np.arange(len(labels))
width=0.5

fig, ax=plt.subplots(figsize=(10,5))
rects=ax.bar(ind,np.array(temp_df['batsman_runs']),width=width)
ax.set_xticklabels(labels,rotation='vertical')
ax.set_xticks(ind+((width)/10))
ax.set_title("Batsman with most number of dot balls",fontweight='bold')

for p in ax.patches:
    ax.annotate(format(p.get_height()),(p.get_x()+0.15,p.get_height()+1))


                     
plt.show()
```

# Bowler Analysis

```python
# Now let us see the bowler who has bowled most number of balls in IPL
temp_df=delivery.groupby('bowler')['ball'].agg('count').reset_index().sort_values(by='ball',ascending=False).reset_index(drop=True)
temp_df=temp_df.iloc[:10,:]

labels=np.array(temp_df['bowler'])
ind=np.arange(len(labels))
width=0.5

fig, ax=plt.subplots(figsize=(10,5))
rects=ax.bar(ind,np.array(temp_df['ball']),width=width)
ax.set_xticks(ind+((width)/2))
ax.set_xticklabels(labels,rotation='vertical')
ax.set_xlabel("Bowler")
ax.set_ylabel("count")
ax.set_title("Number of balls bowled by each bowler",fontweight='bold')

for p in ax.patches:
    ax.annotate(format(p.get_height()),(p.get_x()+0.15,p.get_height()+1))

plt.show()
```

Harbhajan Singh is the bowler who has bowled most number of balls in IPL 

```python
# Now let us see the bowler with more number of dot balls.
temp_df=delivery.groupby('bowler')['total_runs'].agg(lambda x:(x==0).sum()).reset_index().sort_values(by='total_runs',ascending=False).reset_index(drop=True)
temp_df=temp_df.iloc[:10,:]
labels=np.array(temp_df['bowler'])
ind=np.arange(len(labels))
width=0.5

fig, ax=plt.subplots(figsize=(10,5))
rects=ax.bar(ind,np.array(temp_df['total_runs']),width=width)
ax.set_xticklabels(labels,rotation='vertical')
ax.set_xticks(ind+((width)/2))

ax.set_title('Bowler with more number of dot balls',fontweight='bold')

for p in ax.patches:
    ax.annotate(format(p.get_height()),(p.get_x()+0.15,p.get_height()+1))

plt.show()
```

Harbhajan singh is the bowler with more number of dot balls bowled.

```python
# Now let us see the bowler who has given more number of extra runs in IPL
temp_df=delivery.groupby('bowler')['extra_runs'].agg(lambda x:(x>0).sum()).reset_index().sort_values(by='extra_runs',ascending=False).reset_index(drop=True)
temp_df=temp_df.iloc[:10,:]
labels=np.array(temp_df['bowler'])
ind=np.arange(len(labels))
width=0.5

fig, ax=plt.subplots(figsize=(10,5))
rects=ax.bar(ind,np.array(temp_df['extra_runs']),width=width)
ax.set_xticklabels(labels,rotation='vertical')
ax.set_xticks(ind+((width)/2))

ax.set_title('Bowler with more number of extra runs ',fontweight='bold')

for p in ax.patches:
    ax.annotate(format(p.get_height()),(p.get_x()+0.15,p.get_height()+1))

plt.show()
```

Malinga is the bowler who given more extra runs in IPL.

```python
# Highest wicket taker 
plt.subplots(figsize=(10,5))
dismissal_kind=['caught','bowled','stumped','hit-wicket','lbw','caught and bowled']
temp_df=delivery[delivery['dismissal_kind'].isin(dismissal_kind)]
ax=temp_df['bowler'].value_counts()[:10].plot.bar(width=0.5)
ax.set_title("Highest wicket taker",fontweight="bold")
ax.set_xlabel("Bowler")

for p in ax.patches:
    ax.annotate(format(p.get_height()),(p.get_x()+0.15,p.get_height()+1))
plt.show()
```

```python
# Let us see Bowler who bowled maximum overs
temp_df=delivery.groupby(['bowler']).sum()
temp_df['total balls']=delivery['bowler'].value_counts()
temp_df['overs']=(temp_df['total balls']//6)
temp_df[temp_df['overs']>400].sort_values(by='overs',ascending=False)['overs'].reset_index().T
```

```python
# IPL Final Match analysis
finals=matches.drop_duplicates(subset=['season'],keep='last')
finals=finals[['id','season','city','team1','team2','toss_winner','toss_decision','winner']]
most_finals=pd.concat([finals['team1'],finals['team2']]).value_counts().reset_index()
most_finals.rename({'index':'team',0:'count'},axis=1,inplace=True)
xyz=finals['winner'].value_counts().reset_index()
most_finals=most_finals.merge(xyz,left_on='team',right_on='index',how='outer')
most_finals=most_finals.replace(np.NaN,0)
most_finals.drop('index',axis=1,inplace=True)
most_finals.set_index('team',inplace=True)
most_finals.rename({'count':'finals_played','winner':'won_count'},inplace=True,axis=1)
most_finals.plot.bar(width=0.8)
plt.gcf().set_size_inches(10,6)
plt.show()
```
